```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

df=pd.read_csv('homeprices-multiple.csv')
df
```

```python
import math
df.bedrooms=math.floor(df.bedrooms.median())

df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
reg.predict([[3300,3,10]])
```

```python
import matplotlib.pyplot as plt
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='black',marker="*")
```
